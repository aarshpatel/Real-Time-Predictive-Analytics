```python
import pandas as pd
```

```python
df = pd.read_csv("fault_errors_final.csv")
del df['Stud-ID']
del df['Fault #']
del df['Fault Trigger']
del df['Fixed']
del df['Duration']
del df['Reason']
del df['Description']
del df['Extended description']
del df['Operation Mode']
df.head(10)
```

```python
df.iloc[0, 0]
```

```python
error_probs = []
statuses = []
history = []
for idx, row in df.iterrows():
    if len(history) >= 7:
        history.pop(0)
    history.append(df.iloc[idx, 1])
    statuses.append(df.iloc[idx, 1])
    
    total = 0
    for status in history:
        if status == 1:
            total += 1
            
    error_probs.append(float(total)/7)
df["Error_Probs"] = error_probs
```
